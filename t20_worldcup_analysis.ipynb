```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('C:/Users/DELL/Downloads/t20_wc_match_results.json') as f:
    data=json.load(f)
data
```

Data is an array which has this list as one element.
To get this data in tabular and normalised form we will bw using pandas dataframe 

```python
data[0]['matchSummary']
```

```python
df_summary=pd.DataFrame(data[0]['matchSummary'])
df_summary
```

```python
df_summary.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_summary
```

Used rename () function to rename scorecard column to match_id for ease of analysis 

```python
match_ids_dict={}

for index,rows in df_summary.iterrows():
    key1=rows['team1']+ ' Vs ' +rows['team2']
    key2=rows['team2']+ ' Vs ' +rows['team1']
    match_ids_dict[key1]=rows["match_id"]
    match_ids_dict[key2]=rows["match_id"]
match_ids_dict
```

```python
with open ('C:/Users/DELL/Downloads/t20_wc_batting_summary.json') as f:
    data=json.load(f)
    all_record=[]
    for rec in data:
        all_record.extend(rec['battingSummary'])
all_record
```

```python
all_record=pd.DataFrame(all_record)
all_record.head(10)
```

```python
all_record.shape
```

```python
all_record['out/not_out']=all_record.dismissal.apply(lambda x: "OUT" if len(x)>0 else "NOT OUT")
sub=all_record.tail(10)
sub
```

```python
plt.bar(sub['batsmanName'],sub['runs'])
plt.title('Runs by Batsman')
plt.xlabel('Name')
plt.ylabel('Runs')
plt.xticks(rotation=90)
plt.show()
```

To check status of batsman wether he is out or not in a particulat inning i have used 'apply' function and lambda function to iterate throught all elements in list . If there will be any string or lement whos size will bw greater than 0 then that column will be assigned 'out' else 'not out'. 

```python
all_record['dismissal']=all_record.drop(['dismissal'],axis=1,inplace=True)
all_record.head()
```

```python
all_record['batsmanName']=all_record['batsmanName'].apply(lambda a: a.replace('â€',' '))
all_record['batsmanName']=all_record['batsmanName'].apply(lambda a: a.replace('\xa0',' '))

all_record.head(10)
```

```python
df_batting=all_record.drop('dismissal',axis=1)
df_batting
```

```python
df_batting["match_id"]=df_batting["match"].map(match_ids_dict)
df_batting  .head(10)
```

```python
df_batting.to_csv('E:/coding/t20_csv_files/temp.csv',index=0)
```

                                               Processing Bowling data

```python
with open('C:/Users/DELL/Downloads/t20_wc_bowling_summary.json') as f:
    data1=json.load(f)
    all_bowers_record=[]
    for rec in data1:
        all_bowers_record.extend(rec['bowlingSummary'])

all_bowers_record
```

```python
df_bowling=pd.DataFrame(all_bowers_record)
df_bowling.head()
```

```python
df_bowling['match_id']=df_bowling['match'].map(match_ids_dict)
sub1=df_bowling.tail(10)
sub1    
```

```python
plt.bar(sub1['bowlerName'],sub1['wickets'])
plt.title('Bowlers Wicket')
plt.xlabel('Bowler Name')
plt.ylabel('Wickets')
plt.xticks(rotation=90)
plt.show()
```

```python
df_bowling.to_csv('E:/coding/t20_csv_files/bowling.csv',index=0)
```
